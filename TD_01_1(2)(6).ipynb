## 1. Conversion d'une base b vers la base d√©cimale

### 1.1 M√©thode de droite √† gauche

Voici un algorihtme permettant de convertir un nombre entier √©crit en base $b$ vers la base d√©cimale :
> resultat $\leftarrow$ 0<br>
> i $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de droite √† gauche : <br>
>>    resultat $\leftarrow$ resultat + $c \times b^i$<br>
>>    i $\leftarrow$ i+1<br>
>
> Renvoyer resultat<br>

**Q1. Programmer une fonction ``convert_dec`` qui, selon l'algorithme pr√©c√©dent, prend en argument un nombre entier et la base dans laquelle il est √©crit et renvoie sa valeur d√©cimale. Penser √† sp√©cifier la fonction et √©ventuellement √† utiliser ``assert``.**

```python

def convert_dec(c,b):
    ch=str(c)
    result = 0
    i=0
    nb=len(ch)
    for i in range (nb):
        result=result+int(ch[i]-1-i)*b**i
        i=i+1
    return result

print(convert_dec(1111,2))
```

### 1.2 M√©thode de gauche √† droite (m√©thode de Horner)

Voici un 2√®me algorihtme permettant aussi de convertir un nombre entier √©crit en base $b$ vers la base d√©cimale mais en parcourant le nombre dans l'autre sens :
> resultat $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de gauche √† droite : <br>
>>    resultat $\leftarrow$ b $\times$ resultat<br>
>>    resultat $\leftarrow$ resultat + $c$<br>
>
> Renvoyer resultat<br>

**Remarque :** Cette m√©thode peut para√Ætre plus complexe que la premi√®re mais elle est en fait bien plus efficace (rapide) car elle ne n√©cessite que des op√©rations simples (multiplication et addition) au lieu d'utiliser des puissances.

**Q3. Programmer une fonction ``convert_horner`` qui, selon l'algorithme pr√©c√©dent, prend en argument un nombre entier et la base dans laquelle il est √©crit et renvoie sa valeur d√©cimale. Penser √† sp√©cifier la fonction et √©ventuellement √† utiliser ``assert``.**

```python

def convert_dec(c,b):
    ch=str(c)
    result = 0
    i=0
    nb=len(ch)
    for i in range (nb):
        result=b*result
        result=result+int(ch[i])
    return result

print(convert_dec(101001,2))
```

## 2. Conversion la base d√©cimale vers une base b

### 2.1 Conversion par divisions successives par $b$
Pour convertir un nombre entier de la base d√©cimale vers une base $b$, il suffit de le diviser par $b$ (division enti√®re), de noter le reste puis de diviser √† nouveau le r√©sultat par $b$ et de noter le reste et ainsi de suite jusqu'√† obtenir 0. En √©crivant les restes obtenus en commen√ßant par le dernier, on obtient le r√©sultat de la conversion.

**Q5. Programmer une fonction ``convert_base`` qui, selon l'algorithme pr√©c√©dent, prend en argument un nombre entier dans la base d√©cimale et la base dans laquelle on veut le convertir et renvoie la valeur convertie. Penser √† sp√©cifier la fonction et √©ventuellement √† utiliser ``assert``.**

```python
#def convert_base(c,b):
    #div=0
    #reste = 1
    #ch=''
    #while reste>0:
        #ch=ch+str()
        
def convert_base(nb_entier,b):
    """nb_entier est un entier positif et b un entier compris entre 2 et 16.
    Cette fonction renvoie la valeur de nb_entier dans la base b sous forme d'un entier si b<9 et
    d'une cha√Æne de caract√®res sinon. """
    assert type(b) == int, "La base indiqu√©e doit √™tre un nombre entier."
    assert (b>1) and (b<17), "La base indiqu√©e doit √™tre un entier compris entre 2 et 16."
    restes='' # cha√Æne de caract√®re vide
    quotient=nb_entier
    if nb_entier == 0 :
        return nb_entier
    while quotient != 0:
        temp=quotient%b #reste de la divions euclidienne par b
        if temp == 10 : # si le reste est sup√©rieur √† 9, on le remplace par une lettre
            temp = 'A'
        elif temp == 11 :
            temp = 'B'
        elif temp == 12 :
            temp = 'C'
        elif temp == 13 :
            temp = 'D'
        elif temp == 14 :
            temp = 'E'
        elif temp == 15 :
            temp = 'F'
        else :
            temp = str(quotient%b) # si le reste est < 10, on le convertit en cha√Æne de caract√®res
        restes=restes+temp #addtion de cha√Ænes de caract√®res = concat√©nation
        quotient=quotient//b # le prochain quotient est obtenu en divisant par b
    resultat='' #il faut inverser les caract√®res pour le r√©sultat
    for i in range(len(restes)):
        resultat=resultat+restes[len(restes)-i-1] #on ajoute les caract√®res de restes √† resultat en commencant par la fin
    if b<10 :
        return int(resultat) # retourne un entier si la base < 10
    else :
        return resultat # retourne une cha√Æne de caract√®res sinon
```

**Q6. Utiliser la fonction``convert_base`` pr√©c√©dente pour convertir le nombre 213 dans les bases 2, 4 et 16.**

```python
print(convert_base(213,2))
print(convert_base(213,4))
print(convert_base(213,16))
```

### 2.2 Conversion par divisions successives par $b^i$
On utilisera plut√¥t cette m√©thode si on veut fixer le nombre $k$ de chiffres utilis√©s pour √©crire le r√©sultat, quitte √† ce que ce r√©sultat commence par des 0.

Il suffit alors d'effectuer la division enti√®re du nombre √† convertir par $b^{k-1}$ puis de recommencer en divisant le reste par $b^{k-2}$ et ainsi de suite judsqu'√† diviser par $b^0=1$. Les r√©sultats successifs de ces divisions dans l'ordre forment le nombre final. Evidemment, cela ne fonctionne pas si le nombre √† convertir est trop grand pour le nombre de chiffres choisi pour le r√©sultat.

Dans le cas d'une conversion vers le binaire, l'agorithme est le suivant :
> **Pour** $i$ allant de $k-1$ √† $0$ : <br>
>> **Si** $n \geq b^i$ :<br>
>>>    $n \leftarrow n-b^i $<br>
>>>    Rajouter le chiffre $1$ au r√©sultat<br>
>>
>> **Sinon** :<br>
>>> Rajouter le chiffre $0$ au r√©sultat<br>
>
> Renvoyer resultat<br>

**Q7. Programmer une fonction ``convert_bin`` qui, selon l'algorithme pr√©c√©dent, prend en argument un nombre entier dans la base d√©cimale et le nombre de chiffres souhait√© pour le r√©sultat et renvoie la valeur convertie en binaire. Penser √† sp√©cifier la fonction et √©ventuellement √† utiliser ``assert``.**

```python
def convert_bin(nb,e):
    assert (type(e) == int) and (e>0), 
    assert (type(nb) == int) and (nb >0), 
    assert nb<2**e, 
    result=''
    for i in range (e-1,-1,-1): #Pour  ùëñ  allant de  ùëò‚àí1  √†  0  
        if nb>=2**i: #Si  ùëõ‚â•ùëèùëñ
            nb=nb-2**i #ùëõ‚Üêùëõ‚àíùëèùëñ 
            result=result+'1' #Rajouter le chiffre  1  au r√©sultat
        else:
            result=result+'0'
    return result
            
```
